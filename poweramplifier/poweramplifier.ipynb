```python
# setup
import sys
import warnings
import numpy as np
import matplotlib.pyplot as plt
from scipy import special

warnings.filterwarnings("always")
```

### Non-Linear model of BJT transistor

```python
# x is Vac/vt
x = np.linspace(0, 2, 1000)

# h is 2*I1(x)/I0(x)
y = 2 * np.divide(special.iv(1, x), special.iv(0, x))

plt.figure(1)
plt.plot(x, y)
plt.show()
```
