## A Fully Autonomous Ultra-low Power Hybrid RF/Photovoltaic Energy Harvesting System with -25 dBm Sensitivity

The paper [(link)](https://ieeexplore.ieee.org/stamp/stamp.jsp?tp=&arnumber=7953858) proposes a hybrid RF/PV energy harvesting system using the following diagram:

![](img/diagram.png)

For the RF-DC converter 3 rectifiers were studied:

1. Series rectifier
2. One-stage voltage multiplier
3. Two-stage voltage multiplier

The comparison can be seen on the table:

![](img/table1.png)

The rectifier used was a series one due to better efficiency (measured via ADS simulation).

To amplify the DC voltage generated by the rectifier (which is farily low) a boost DC-DC converter was used. The picture below presents the architecure:

![](img/boost.png)

- The capacitor Cd was added to extend the oscillator working time to 60 ms;
- The timer has a period of 16 seconds, then it triggers the DC-RF conversion;
- An output voltage of 1.5 V was set as goal because it is the timer minimum supply and the highest required voltage of the entire circuit;
- The oscillator works at 10 kHz to improve its performance without degrading too much the boost efficiency.

The table below presents the consumption of the timer and the oscillator. The timer works for 31.25 ms and the oscillator for another 30 ms due to the presence of Cd.

![](img/table2.png)

The last image presents the prototype and the waveform of the voltages in the system during a long period of time. The RF-DC performance achieved was in average 17.6% and the voltage in the boost converter was above 1.5 V.
